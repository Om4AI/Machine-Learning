## ***Fifa dataset***

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fifa_data = pd.read_csv("fifa.csv", index_col="Date", parse_dates=True)
print(type(fifa_data))
print("Shape: ",fifa_data.shape)
fifa_data.head()
```

```python
# First plot
# Matplotlib to set up the figure
plt.figure(figsize=(16,6))

# Plot a line graph for the Fifa Dataset
sns.lineplot(data=fifa_data)
```

## ***Bar Charts & Heat Maps: Flight Delays dataset*** 

```python
flight_data = pd.read_csv("flight_delays.csv", index_col="Month")
print(flight_data.shape)
```

```python
# Explore data
flight_data
```

```python
print(flight_data["AA"].mean())
print(flight_data["NK"].mean())
```

```python
flight_data["AA"].max()
```

### ***Barchart: Spirit Airlines (Code: NK)***

#### ***Barplot for full dataset***

```python
plt.figure(figsize=(14,6))

sns.set_style("darkgrid")

sns.barplot(data = flight_data)
plt.xlabel("Airlines")
plt.ylabel("Delays")
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.title("Average arrival delay by Spirit Airlines Flights monthwise")

sns.barplot(x= flight_data.index, y=flight_data['NK'])

plt.ylabel("Arrival Delay (in minutes)")
plt.show()
```

### ***HeatMap:For full dataset***

```python
plt.figure(figsize=(14,7))

plt.title("Average arrival delay for airlines: Heatmap")
# annot-True ensures that the values appear on the heatmap
sns.heatmap(data=flight_data, annot=True)

plt.xlabel("Airlines")
```

#### ***The white line in "US" airlines is due to the NaN values for those months***

## ***Scatter Plots: Insurance Dataset***

```python
# Get data

insurance_data = pd.read_csv("insurance.csv")
insurance_data.head()
```

```python
insurance_data.shape
```

```python
# Scatter plot

sns.scatterplot(x = insurance_data["bmi"], y=insurance_data["charges"])
```

```python
# Getting a regression line on the data

sns.regplot(x = insurance_data["bmi"], y=insurance_data["charges"])
```

### ***Implementating Color codes*** 

```python
sns.scatterplot(x=insurance_data["bmi"], y=insurance_data["charges"], hue=insurance_data["smoker"])
```

```python
# Get 2 regression lines for different categories

sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

### ***Categorical scatter plot***

```python
sns.swarmplot(x=insurance_data["smoker"], y=insurance_data["charges"])
```

## ***Distributions: Histograms & Density Plots***

### ***Histograms***

```python
# Get data

iris_data = pd.read_csv("iris.csv", index_col="Id")

iris_data.head()
```

```python
sns.distplot(a=iris_data["Petal Length (cm)"], kde=False)
```

### ***Density Plots***

#### ***Single Dimensional KDE***

```python
sns.kdeplot(data = iris_data["Petal Length (cm)"], shade=True)
```

#### ***2D KDE plots***

```python
# 2D KDE plot
# This helps us to plot the KDE for more than 1 columns

sns.jointplot(x = iris_data['Petal Length (cm)'], y = iris_data["Sepal Width (cm)"], kind="kde", shade=True)
```

### ***Color-Coded plots: Histograms & KDE plots***

```python
iris_set_data = pd.read_csv("iris_setosa.csv", index_col="Id")
iris_ver_data = pd.read_csv("iris_versicolor.csv", index_col="Id")
iris_vir_data = pd.read_csv("iris_virginica.csv", index_col="Id")

iris_ver_data.head()
```

### ***Histogram for all the 3 species***

```python
plt.figure(figsize=(13,7))

sns.distplot(a=iris_set_data["Petal Length (cm)"], label="Iris Setosa", kde=False)
sns.distplot(a=iris_ver_data["Petal Length (cm)"], label="Iris Versicolor", kde=False)
sns.distplot(a=iris_vir_data["Petal Length (cm)"], label="Iris Virginica", kde=False)

plt.title("Histogram of Petal Length, by species")

# Legend
plt.legend()
```

### ***KDE Plots for each species***

```python
plt.figure(figsize=(13,7))

sns.kdeplot(data =iris_set_data["Petal Length (cm)"], label="Iris Setosa", shade=True)
sns.kdeplot(data =iris_ver_data["Petal Length (cm)"], label="Iris Versicolor", shade=True)
sns.kdeplot(data =iris_vir_data["Petal Length (cm)"], label="Iris Virginica", shade=True)

plt.title("Distribution of species using KDE plot")
```
